```python
import numpy as np
import matplotlib.pyplot as plt
import astropy.io.fits as fits
```

```python
dat_file = '/Users/cmason/Downloads/combined_fluxcalib_abs_slits_lin.fits'
hdu_list = fits.open(dat_file)
```

```python
obj_id = 144

dat2D = hdu_list[obj_id].data
dat1D_med = np.nanmedian(dat2D, axis=0)

ypos = hdu_list[obj_id].header['SLITYPIX']
print(ypos)

y_width_pix = 8


plt.figure(figsize=(10, 3), dpi=150)

dat1D = np.nansum(dat2D[int(ypos-y_width_pix/2.):int(ypos+y_width_pix/2.)+1], axis=0)
plt.plot(dat1D, lw=1, c='0.6')

dat1D_medsub = np.nansum((dat2D-dat1D_med)[int(ypos-y_width_pix/2.):int(ypos+y_width_pix/2.)+1], axis=0)
plt.plot(dat1D_medsub, lw=1)

plt.xlim(3000, 5000)
# plt.ylim(-0.5e-18, 0.75e-18)
```
